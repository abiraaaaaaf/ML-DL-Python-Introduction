#Functions
###User Defined Functions
Functions that we define ourselves to do certain specific task are referred as user-defined functions. In Python, a user-defined function's declaration begins with the keyword def and followed by the function name. The function may take arguments(s) as input within the opening and closing parentheses, just after the function name followed by a colon. 


```
def function(arguments):
  <statement>
  ...
  <statement>
```


````python
def greet(name):
    """
    This function greets to
    the person passed in as
    a parameter
    """
    print("Hello, " + name + ". Good morning!")
````

````python
greet("Ali")
````


```
def function(inputs):
  <statement>
  ...
  <statement>
  return outputs
  ```


````python
def add(a,b):
  c = a + b 
  return c
````

````python
add(6,8)
````

````python
def absolute_value(num):
    """This function returns the absolute
    value of the entered number"""

    if num >= 0:
        return num
    else:
        return -num
````

###Load *iris* Dataset

````python
import sklearn
````

````python
from sklearn.datasets import load_iris
iris = load_iris()
````

````python
X = iris.data   # samples
y = iris.target  # labels
````

````python
X
````

````python
print(len(X[2]))
````

````python
y
````

````python
print(len(y))
````

````python
feature_names = iris.feature_names
target_names = iris.target_names
print("Feature names:", feature_names)
print("Target names:", target_names)
print("\nFirst 10 rows of X:\n", X[:10])
````

###Splitting the dataset
Link *train_test_split()*: https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html

train-and-test-1-min-1.png

Cross-Validation-In-Machine-Learning-1.jpg

````python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
   X, y, test_size = 0.3, random_state = 1)

print(X_train.shape)
print(X_test.shape)

print(y_train.shape)
print(y_test.shape)
````

###Train the Model


KNN_final1_ibdm8a.png

###k-nearest neighbors algorithm (KNN)
ss.JPG

````python
from sklearn.neighbors import KNeighborsClassifier
classifier_knn = KNeighborsClassifier(n_neighbors = 11)
classifier_knn.fit(X_train, y_train)
````

- Finding accuracy by comparing actual response values(y_test) with predicted response value(y_pred)


````python
from sklearn import metrics
y_pred = classifier_knn.predict(X_test)
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
````

````python
X_test
````

- Providing sample data and the model will make prediction out of that data

````python
sample = [[5, 5, 3, 2], [2, 4, 3, 5], [2, 5, 6, 9]]
preds = classifier_knn.predict(sample)
pred_species = [iris.target_names[p] for p in preds] 
print("Predictions:", pred_species)
````

###Save the Model

````python
import joblib
joblib.dump(classifier_knn, 'iris_classifier_knn.joblib')
````
